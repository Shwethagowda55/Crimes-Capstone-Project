```python
Literacy_Rate.head()
```

```python
Literacy_Rate.tail()
```

```python
# change the column name to State/ UT

Literacy_Rate.rename(columns={"State/ UTs": "State/ UT"},inplace=True)
```

```python
Literacy_Rate.head()
```

```python
columns_to_remove = ['1951', '1961', '1971', '1981', '1991']
```

```python
Literacy_Rate.drop(columns=columns_to_remove, inplace=True)
```

```python
Literacy_Rate.rename(columns={"2001": "Literacy_Rate 2001","2011":"Literacy_Rate 2011"},inplace=True)
```

```python
Literacy_Rate.head()
```

```python
#Removed '1951', '1961', '1971', '1981', '1991 years literacy rate because we are looking for only 2001 and 2011.
```

```python
Literacy_Rate.shape
```

```python
Literacy_Rate.columns
```

```python
Literacy_Rate.dtypes
```

```python
Literacy_Rate.info()
```

```python
Literacy_Rate.isnull().sum()
```

```python
# lets visualize it using heatmap

sns.heatmap(Literacy_Rate.isnull())
```

```python
# Assuming 'Literacy_Rate' is your DataFrame
# Replace this name with your actual DataFrame

# Fill missing values in '2001' column with 60.47
Literacy_Rate['Literacy_Rate 2001'].fillna(60.47, inplace=True)

# Fill missing values in '2011' column with 60.47
Literacy_Rate['Literacy_Rate 2011'].fillna(60.47, inplace=True)
```

```python
Literacy_Rate.isnull().sum()
```

```python
Literacy_Rate.plot(kind='bar', figsize=(12, 8))
plt.title('Literacy Rate Comparison (2001 vs. 2011)')
plt.xlabel('State')
plt.ylabel('Literacy Rate')
plt.legend(["Literacy Rate 2001", "Literacy Rate 2011"])
plt.show()
```
